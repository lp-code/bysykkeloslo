```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_feather('feature_importances.feather')
```

```python
df.describe()
```

```python
medians = df.median().nlargest(10)
#medians.sort_values(ascending=False, inplace=True)
df = df[medians.index]
```

```python
ax = sns.boxplot(data=df, orient='h')
_ = ax.set_yticklabels([lbl.get_text().replace('_', ' ') for lbl in ax.get_yticklabels()])
_ = ax.set_xlabel('feature importance (statistics over individual models for all stations)')
```

```python
ax.get_figure().savefig('feature_importances.svg')
```
